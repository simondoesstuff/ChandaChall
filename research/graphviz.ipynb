```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import graphviz

from io_utils import import_data, import_solution
```

```python
n = 1109

dot = graphviz.Digraph()
soln = import_solution('../solutions/naive/' + str(n) + '.soln')
alias = dict()
id = 1

for parent in soln:
    for child in soln[parent]:
        if parent not in alias:
            alias[parent] = id
            id += 1
        if child not in alias:
            alias[child] = id
            id += 1
        
        parent_alias = alias[parent]
        child_alias = alias[child]         
        dot.edge(str(parent_alias), str(child_alias))

print(dot.source)
dot.render(view=True)
```
