```python
import io_utils
```

```python
d = 3515

naive = io_utils.import_solution(f'../solutions/naive/{d}.soln')
trickle_threads = io_utils.import_solution(f'../solutions/trickle_threads/{d}.soln')
trickle_sync = io_utils.import_solution(f'../solutions/trickle_sync/{d}.soln')
```

```python
file = f"../solutions/slb_rust/{d}.soln"
with open(file) as f:
    edges = dict()
    
    for line in f:
        (parent, child) = line.split('->')
        to_node = lambda x: tuple(sorted([int(e) for e in x.strip().split(' ')]))
        parent_labels = to_node(parent)
        child_labels = to_node(child)
        edges.setdefault(parent_labels, set())
        edges[parent_labels].add(child_labels)
    
    rust = edges
```

```python
def compare(contendor):
    print(f"Contendor keys are {rust.keys()}")
    print(f"Naive keys are {naive.keys()}")
    
    for key in naive:
        if key not in contendor:
            print('missing key: {}'.format(key))
        elif naive[key] != contendor[key]:
            print('mismatch: {} -> {} != {} -> {}'.format(key, naive[key], key, contendor[key]))
```

# Comparison

```python
compare(rust)
```

```python
compare(trickle_threads)
```

```python
compare(trickle_sync)
```
